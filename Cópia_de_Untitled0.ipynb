# Análise de Desempenho Escolar com Python

## Contexto
Este projeto utiliza dados fictícios de uma escola para analisar o desempenho acadêmico dos alunos.

## Objetivo
Aplicar técnicas de análise de dados para apoiar decisões pedagógicas, identificando turmas e alunos que necessitam de acompanhamento.

## Tecnologias Utilizadas
- Python
- Pandas
- Matplotlib
- Google Colab




## Entendimento dos Dados

O conjunto de dados representa registros acadêmicos de alunos, contendo
informações sobre turma, disciplina, bimestre, nota e frequência.
Cada linha corresponde ao desempenho de um aluno em um determinado bimestre.


### Descrição das Variáveis

- **aluno_id**: identificador do aluno  
- **turma**: turma do aluno  
- **disciplina**: disciplina cursada  
- **bimestre**: período letivo  
- **nota**: nota obtida pelo aluno  
- **frequencia**: percentual de presença


```python
media_turma = df.groupby('turma')['nota'].mean()
media_turma
```

“A análise das médias por turma permite identificar quais grupos apresentam maior dificuldade de aprendizagem, possibilitando intervenções pedagógicas mais direcionadas.”

```python
media_disciplina = df.groupby('disciplina')['nota'].mean()
media_disciplina
```

“Ao analisar as médias por disciplina, é possível identificar áreas do conhecimento que exigem maior atenção dos alunos.”

```python
import matplotlib.pyplot as plt

media_turma.plot(kind='bar')
plt.title('Média de Notas por Turma')
plt.xlabel('Turma')
plt.ylabel('Nota Média')
plt.show()
```

```python
media_disciplina.plot(kind='bar')
plt.title('Média de Notas por Disciplina')
plt.xlabel('Disciplina')
plt.ylabel('Nota Média')
plt.show()
```

“Os gráficos facilitam a visualização do desempenho acadêmico, permitindo identificar rapidamente turmas ou disciplinas que necessitam de ações pedagógicas específicas.”
